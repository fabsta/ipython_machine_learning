##Entropy
Where to split the data
Definition: measure of impurity in a bunch of examples

$$entropy = \sum{-p_i * log (p_i)}$$

Ranges from 0 (all samples from 1 class) to 1 (samples split across classes) 

```python
#Calculating the entropy for 

import math

-0.5*math.log(0.5,2) +(- 0.5*math.log(0.5,2))

-0.66*math.log(0.66,2) +(-0.33*math.log(0.33,2))
```

###Information gain
information gain = entropy(parent) - [weighted avg] entropy (children)

decision tree algorithm -> maximize information gain

```python
#information gain
# for bumpiness
entropy_parent = 1
entropy_children =  float(3./4.)* (0.9184)+float(1./4.)*0

entropy_parent - entropy_children
```
